```python
#Brute Force

def solve(height):
    n=len(height)
    maxwater=0

    for i in range(n):
        for j in range(i+1,n):
            width=j-i
            water=width*min(height[i],height[j])
            maxwater=max(maxwater,water)
    return maxwater
height = [1,8,6,2,5,4,8,3,7]
solve(height)
```

```python
# Optimal approach ( Two Pointer)

def solve(h):
    n=len(h)
    left=0
    right=n-1
    maxwater=0

    while left<right:
        width=right-left
        water=width*min(h[left],h[right])
        maxwater=max(maxwater,water)
    
        if h[left]<h[right]:
            left+=1
        else:
            right-=1
    return maxwater

height = [1,8,6,2,5,4,8,3,7]
solve(height)
```
