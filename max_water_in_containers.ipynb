```python

import numpy as np

def max_water_bruteforce(heights):
    heights=np.array(heights)
    n=len(heights)
    max_water=0

    for i in range(n):
        for j in range(i+1,n):
            width=j-i
            area=min(heights[i],heights[j])*width
            max_water=max(area,max_water)

    return max_water

heights=[1,8,6,2,5,4,8,3,7]
print(max_water_bruteforce(heights))
```

```python
def max_water(heights):
    n=len(heights)
    max_water_ans=0;

    left,right=0,n-1

    while left < right:
        area=min(heights[left],heights[right])*(right-left)
        max_water_ans=max(max_water_ans,area)
        if(heights[left]<heights[right]):
            left+=1
        else:
            right-=1

    return max_water_ans

heights=[1,8,6,2,5,4,8,3,7]
print(max_water(heights))
```
